# Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_boston
```

## Linear regression for house prices in Boston based on the number of rooms

```python
boston_dataset = load_boston()
print(boston_dataset.DESCR)
```

```python
# Features 
X = np.array(boston_dataset.data[:, 5])
# Target
Y = np.array(boston_dataset.target)
```

```python
# Plot the selected data
plt.scatter(X, Y, alpha=0.3)
plt.xlabel('average number of rooms per dwelling')
plt.ylabel('house-price')
plt.show()
```

```python
# A column of ones is added for the independent term
X = np.array([np.ones(X.shape[0]), X]).T
```

Minimization of the mean square error:
$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
B = np.linalg.inv(X.T @ X) @ X.T @ Y
```

```python
# Plot results
plt.scatter(X.T[1], Y, alpha=0.3)
min_plot_x, max_plot_x = 4, 9
plt.plot([min_plot_x, max_plot_x], [B[0] + B[1] * min_plot_x, B[0] + B[1] * max_plot_x], c='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel('house-price')
plt.show()
```
