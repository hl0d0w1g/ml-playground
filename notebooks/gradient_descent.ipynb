# Gradient Descent 

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Function definition and visualization

$F(x, y) = sin(\frac{1}{2}x^2 - \frac{1}{4}y^2 + 3)cos(2x + 1 - e^y)$

```python
func = lambda th: np.sin(((1 / 2) * (th[0] ** 2)) - ((1 / 4) * (th[0] ** 2)) + 3) * np.cos((2 * th[0]) + 1 - (np.e ** th[1]))
```

```python
res = 100
x = np.linspace(-2, 2, res)
y = np.linspace(-2, 2, res)
x, y = np.meshgrid(x, y)
z = func([x, y])
```

```python
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(221, projection='3d')
surface_plot = ax1.plot_surface(x, y, z, cmap='winter')

ax2 = fig.add_subplot(222)
contour_plot = ax2.contourf(x, y, z, 100, cmap='winter')
plt.colorbar(contour_plot, ax=ax2)

plt.show()
```

## Gradient descent calculation

```python
# Set random start point
theta = (np.random.rand(2) * 4) - 2

plt.contourf(x, y, z, 100, cmap='winter')
plt.colorbar()
plt.plot(theta[0], theta[1], 'o', c='red')
plt.show()
```

```python
# Derivative ratio
h = 0.001
# Learning rate
lr = 0.001
# Set a gradient variable
gradient = np.zeros(len(theta))

# Plot contour and start point
plt.contourf(x, y, z, 100, cmap='winter')
plt.colorbar()
plt.plot(theta[0], theta[1], 'o', c='white')

_theta = np.copy(theta)
for _ in range (100000):
    for it, th in enumerate(theta):
        _theta = np.copy(theta)
        # Derivative calculation
        _theta[it] = _theta[it] + h
        deriv = (func(_theta) - func(theta)) / h
        # Update gradient
        gradient[it] = deriv
    
    # Update theta with the learning rate
    theta = theta - (lr * gradient)
    
    if (_ % 100 == 0):
        # Plot some intermediate points
        plt.plot(theta[0], theta[1], '.', c='red') 
    
    
# Plot final point
plt.plot(theta[0], theta[1], 'o', c='black')
plt.show()
```
